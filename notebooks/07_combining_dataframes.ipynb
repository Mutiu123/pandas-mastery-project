# 🔗 Notebook 07: Combining DataFrames## Join and merge data- Merge (SQL-like joins)- Concat (stacking)- Join operations- Handling conflicts

```python
import pandas as pdimport numpy as npemployees = pd.read_csv('../datasets/employees.csv')orders = pd.read_csv('../datasets/orders.csv')customers = pd.read_csv('../datasets/customers.csv')products = pd.read_csv('../datasets/products.csv')
```

## 1. Merge - SQL-Style Joins

```python
# Inner joinmerged = pd.merge(    orders,    customers,    on='customer_id',    how='inner')print(f'Inner join result: {len(merged)} rows')
```

```python
# Left joinmerged = pd.merge(    orders,    products,    on='product_id',    how='left')print(merged.head())
```

## 3. Concat - Stacking DataFrames

```python
# Vertical stacking (rows)df1 = employees.head(100)df2 = employees.tail(100)stacked = pd.concat([df1, df2], ignore_index=True)print(f'Stacked: {len(stacked)} rows')
```

```python
# Horizontal stacking (columns)# Be careful with index alignment!df_left = employees[['employee_id', 'first_name']]df_right = employees[['salary', 'department']]combined = pd.concat([df_left, df_right], axis=1)print(combined.head())
```

## 4. Join on Index

```python
# Set index firstdf1 = employees.set_index('employee_id')df2 = employees[['employee_id', 'bonus']].set_index('employee_id')# Joinjoined = df1.join(df2, rsuffix='_bonus')print(joined.head())
```

## 5. Merge with Indicators

```python
# See where data came frommerged = pd.merge(    orders.head(100),    customers,    on='customer_id',    how='outer',    indicator=True)print(merged['_merge'].value_counts())
```
