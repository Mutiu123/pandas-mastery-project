## 4. Pivot Tables

```python
# Create pivot tablepivot = df.pivot_table(    values='salary',    index='department',    columns='position',    aggfunc='mean',    fill_value=0)print(pivot)
```

```python
# With margins (totals)pivot = df.pivot_table(    values='salary',    index='department',    columns='position',    aggfunc='mean',    margins=True)print(pivot)
```

## 5. Crosstab

```python
# Cross-tabulationcross = pd.crosstab(    df['department'],    df['position'],    margins=True)print(cross)
```

```python
# With valuescross_salary = pd.crosstab(    df['department'],    df['position'],    values=df['salary'],    aggfunc='mean')print(cross_salary)
```
