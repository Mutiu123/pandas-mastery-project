## 3. Boolean Indexing - The Power Tool!

```python
# Single conditionhigh_earners = df[df['salary'] > 100000]print(f'High earners: {len(high_earners)}')
```

```python
# Multiple conditions with &, |senior_engineers = df[    (df['department'] == 'Engineering') &     (df['age'] > 40)]print(f'Senior engineers: {len(senior_engineers)}')
```

```python
# NOT condition with ~not_sales = df[~df['department'].isin(['Sales'])]print(f'Non-sales employees: {len(not_sales)}')
```
