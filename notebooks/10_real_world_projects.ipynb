## Project 2: Customer Churn Analysis### Business Question:*"Which customer segments are at risk of churning?"*

```python
customers = pd.read_csv('../datasets/customers.csv')orders = pd.read_csv('../datasets/orders.csv')# Prepare datacustomers['signup_date'] = pd.to_datetime(customers['signup_date'])customers['days_since_signup'] = (pd.Timestamp.now() - customers['signup_date']).dt.days
```

```python
# Define churn (no activity + low lifetime value)customers['at_risk'] = (    (customers['is_active'] == False) |    (customers['total_purchases'] == 0) |    (customers['lifetime_value'] < 500))print(f"Customers at risk: {customers['at_risk'].sum()} ({customers['at_risk'].mean()*100:.1f}%)")
```

```python
# Segment analysissegment_risk = customers.groupby('customer_segment')['at_risk'].agg(['sum', 'mean'])segment_risk.columns = ['At_Risk_Count', 'At_Risk_Rate']print('\nRisk by Segment:')print(segment_risk)
```

```python
# Age analysisage_groups = pd.cut(customers['age'], bins=[0, 30, 45, 60, 100], labels=['18-30', '31-45', '46-60', '60+'])age_risk = customers.groupby(age_groups)['at_risk'].mean()print('\nRisk by Age Group:')print(age_risk)
```

## Project 3: HR Analytics Report### Business Question:*"How can we improve employee retention and satisfaction?"*

```python
employees = pd.read_csv('../datasets/employees.csv')employees['hire_date'] = pd.to_datetime(employees['hire_date'])employees['tenure_years'] = (pd.Timestamp.now() - employees['hire_date']).dt.days / 365
```

```python
# Key metricsprint('=== HR DASHBOARD ===')print(f"Total Employees: {len(employees)}")print(f"Average Salary: ${employees['salary'].mean():,.2f}")print(f"Average Tenure: {employees['tenure_years'].mean():.1f} years")print(f"Average Performance: {employees['performance_score'].mean():.2f}/5")
```

```python
# Department analysisdept_analysis = employees.groupby('department').agg({    'employee_id': 'count',    'salary': 'mean',    'age': 'mean',    'performance_score': 'mean',    'tenure_years': 'mean'}).round(2)print('\nDepartment Analysis:')print(dept_analysis)
```
