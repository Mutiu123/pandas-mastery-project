# ⏰ Notebook 08: Time Series Operations## Work with temporal data- DateTime operations- Resampling- Rolling windows- Time-based filtering

```python
import pandas as pdimport numpy as npimport matplotlib.pyplot as pltfinancial = pd.read_csv('../datasets/financial_data.csv')financial['date'] = pd.to_datetime(financial['date'])financial = financial.set_index('date')
```

## 1. DateTime Basics

```python
employees = pd.read_csv('../datasets/employees.csv')employees['hire_date'] = pd.to_datetime(employees['hire_date'])# Extract componentsemployees['hire_year'] = employees['hire_date'].dt.yearemployees['hire_month'] = employees['hire_date'].dt.monthemployees['hire_day_of_week'] = employees['hire_date'].dt.day_name()print(employees[['hire_date', 'hire_year', 'hire_month']].head())
```

## 2. Time-Based Selection

```python
# Select by yeardata_2024 = financial['2024']print(f'2024 data: {len(data_2024)} days')
```

```python
# Date rangeq1_2024 = financial['2024-01':'2024-03']print(f'Q1 2024: {len(q1_2024)} days')
```

## 3. Resampling

```python
# Daily to weeklyweekly = financial['revenue'].resample('W').sum()print(weekly.head())
```

```python
# Daily to monthlymonthly = financial.resample('M').agg({    'revenue': 'sum',    'expenses': 'sum',    'profit': 'sum'})print(monthly.head())
```

## 4. Rolling Windows

```python
# 7-day rolling averagefinancial['revenue_7d_avg'] = financial['revenue'].rolling(window=7).mean()# 30-day rolling sumfinancial['revenue_30d_sum'] = financial['revenue'].rolling(window=30).sum()print(financial[['revenue', 'revenue_7d_avg', 'revenue_30d_sum']].tail())
```

## 5. Time Differences

```python
# Calculate tenureemployees['tenure_days'] = (pd.Timestamp.now() - employees['hire_date']).dt.daysemployees['tenure_years'] = employees['tenure_days'] / 365print(employees[['hire_date', 'tenure_years']].head())
```

## 6. Plotting Time Series

```python
# Plot revenue over timeplt.figure(figsize=(12, 6))financial['revenue'].plot(label='Daily Revenue')financial['revenue_7d_avg'].plot(label='7-Day Average', linewidth=2)plt.title('Revenue Over Time')plt.xlabel('Date')plt.ylabel('Revenue ($)')plt.legend()plt.grid(True, alpha=0.3)plt.tight_layout()plt.show()
```
